# Logistic Regression (Binary Classification)
### Logistic Regression is a machine learning algorithm based on supervised learning
### It is a statistical method that is used for predicting probability of target variable
### It makes probability for classification problems that are discrete in nature.
### Examples: English or Hindi, True or False, 1 or 0, Right or Wrong, Cat or Dog, etc.

## Two types of Logistic Regression
### Binary Classification, Ex- Win or loss
### Multiclass Classification, Ex- Cat or Dog or Goat or Sheep

## Binary Classification

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
dataset = pd.read_csv(r'E:\arya\machine learning python\logisticreg_dataset.csv')
```

```python
dataset
```

```python
dataset['bought_insurance'].replace({'no':'0','yes':'1'},inplace= True)
```

```python
dataset
```

```python
plt.scatter(x = 'age',y = 'bought_insurance',data = dataset)
```

```python
from sklearn.model_selection import train_test_split
```

```python
#creating testing and training data from dataset
x_train,x_test,y_train,y_test = train_test_split(dataset[['age']],dataset['bought_insurance'],test_size=0.2)
```

```python
len(x_train)
```

```python
len(x_test)
```

```python
#logistic regression
from sklearn.linear_model import LogisticRegression
```

```python
lr = LogisticRegression()
```

```python
lr.fit(x_train,y_train)
```

```python
lr.predict(x_test)
```
